```python
schools = pd.read_csv("../4_Map/geo_files/schools.csv", encoding = "utf8")
with open("../4_Map/geo_files/geojson_districts") as data_file:
    geojson_districts = json.load(data_file, encoding = "utf8")
```

```python
app = dash.Dash()

app.layout = dcc.Graph(
    id = "mapbox",
    figure={
        "data": [
            dict(
                type = "scattermapbox",
                lat = schools["Latitud"],
                lon = schools["Longitud"],
                mode = "markers",
                marker = dict(size = 8, color = "#000000"),
                hoverinfo = "text",
                text = schools["Admisiones_p"]
            )
        ],
        "layout": dict(
            autosize = True,
            hovermode = "closest",
            margin = dict(l = 0, r = 0, t = 0, b = 0),
            legend = dict(
                xanchor = "right",
                x = 0.95,
                yanchor = "top",
                y = 0.9,
                font = dict(
                    #family = "sans-serif",a
                    size = 14,
                    color = "#aaaaaf"
                )
            ),
            mapbox = dict(
                accesstoken = mapbox_access_token,
                bearing = 0,
                center = dict(lat = 40.477, lon = -3.72),
                style = "light",
                pitch = 0,
                zoom = 10.2,
                layers = [
                    dict(
                        type = "fill",
                        sourcetype = "geojson",
                        source = geojson_districts,
                        color = "#eef5db",
                        opacity = 0.8,
                        below = "state-label-sm"
                    )
                ]
            )
        )
    },
    style = {"height": "100vh"}
)

app.run_server()
```
